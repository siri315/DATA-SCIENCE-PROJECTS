# Overview
‘Denco Corp’ a multi-utility manufacturing company sells finished and semi-finished
utility devices and part components to customers across the continent. ‘Denco Corp’s’
customer base consists mainly of small and medium sized businesses that further
process the purchased parts before it becomes a consumer end product. ‘Denco Corp’
relies on repeated sales and higher margin from its customers to expand the business.

# Dataset Description
The “CustomerPartSales” has 7583 rows and 6 columns. The data contains the
simulated sales transaction data of “Denco Corp” a multi-utility manufacturing
company from sales of its parts to various customers. Each row of the transaction data
contain the following 5 fields:
n CUSTNAME: Name of the customer
n REGION: Geographic region class of the customer
n PARTNUM: Part number purchased
n REVENUE: Revenue from the transaction
n COST: Cost to company
n MARGIN: Profit margin from the transaction

# 2. REGION-WISE ANALYSIS

### Let us analize from which region we are getting highest revenue and highest margin 

```python
df.head()
```

```python
# Grouping the region to get unique customers from each region 
unique_cust = df[['REGION','CUSTNAME']].groupby('REGION').nunique().sort_values('CUSTNAME',ascending=False)
```

```python
# Creating a DataFrame of unique cust for future purpose
unique_cust = pd.DataFrame(unique_cust)
unique_cust.reset_index(level=0, inplace=True) 
```

```python
unique_cust.head()
```

‘Denco Corp’ has more customers from East and Central region

 But from which region ‘Denco Corp’  has highest revenue and margin?

```python
# Now again grouping region and summing revenue of every individual region to get more clarity on which region has highest revenue and maring
region_max = df[['REGION','REVENUE','MARGIN']].groupby('REGION').sum().sort_values('REVENUE',ascending=False)
```

```python
# Creating a DataFrame of region with their sum of individual region revenues and margin 
region_max = pd.DataFrame(region_max)
region_max.reset_index(level=0, inplace=True) 
```

```python
# Let us merge Top_region and unique_cust to get more clarity 
Top_region = region_max.merge(unique_cust)
```
